```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('NYC_DF_BIGGEST.csv')
```

```python
X_train = df.drop('FULLVAL',axis=1)
```

```python
fig,ax=plt.subplots(ncols=13,nrows=1,figsize=(25,10),dpi=400)
i=0

for column in X_train.columns:

    sn.boxplot(data=X_train ,ax=ax[i])
    i=i+1
    
plt.tight_layout()
```

```python
for column in X_train.columns:
    sn.kdeplot(data=X_train,x=column)
    plt.show()
```
