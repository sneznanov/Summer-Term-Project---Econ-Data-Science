```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('data.csv')
len(df)
df = df.reset_index()
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(1, inplace=True)
```

```python
X_train = df.drop('MEDV',axis=1)
```

```python
fig,ax=plt.subplots(ncols=14,nrows=1,figsize=(25,10),dpi=400)
i=0

for column in X_train.columns:

    sn.boxplot(data=X_train,y=column,ax=ax[i])
    i=i+1
    
plt.tight_layout()
```

```python
for column in X_train.columns:
    sn.kdeplot(data=X_train,x=column)
    plt.show()
    
```
